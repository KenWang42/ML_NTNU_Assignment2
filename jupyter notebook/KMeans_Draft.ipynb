```python
# get the prob. of each cluster
def setLabel(y, y_res, y_clf, n_clusters):
    prob = []
    for i in range(n_clusters):
        prob.append(y[y_clf == i].sum() / y[y_clf == i].size)
        if prob[i] >= 0.5:
            y_res[y_clf == i] = 1
    print(prob)
```
